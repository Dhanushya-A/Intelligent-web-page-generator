```python
import gradio as gr
from transformers import pipeline
```

```python
generator = pipeline("text-generation", model="gpt2")
```

```python
def generate_webpage(title, description):

    html_output = f"""
    <!DOCTYPE html>
    <html>
    <head>
        <title>{title}</title>
        <style>
            body {{
                font-family: Arial;
                margin: 0;
                background-color: #f4f4f4;
            }}
            header {{
                background: #333;
                color: white;
                padding: 20px;
                text-align: center;
            }}
            section {{
                padding: 20px;
                margin: 20px;
                background: white;
                border-radius: 8px;
            }}
            footer {{
                background: #333;
                color: white;
                text-align: center;
                padding: 10px;
            }}
        </style>
    </head>

    <body>

        <header>
            <h1>{title}</h1>
        </header>

        <section>
            <h2>About</h2>
            <p>{description}</p>
        </section>

        <section>
            <h2>Features</h2>
            <ul>
                <li>Responsive design</li>
                <li>Modern layout</li>
                <li>Easy to customize</li>
            </ul>
        </section>

        <section>
            <h2>Contact</h2>
            <p>Email: example@email.com</p>
        </section>

        <footer>
            <p>© 2026 My Website</p>
        </footer>

    </body>
    </html>
    """

    with open("generated_page.html", "w") as file:
        file.write(html_output)

    return html_output
```

```python
interface = gr.Interface(
    fn=generate_webpage,
    inputs=[
        gr.Textbox(label="Website Title"),
        gr.Textbox(label="Website Description")
    ],
    outputs=gr.Textbox(label="Generated HTML Code"),
    title="Intelligent Web Page Generation System",
    description="Enter a website title and description to generate a webpage using AI."
)
```

```python
interface.launch()
```
